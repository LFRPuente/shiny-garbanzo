# Temporal Series Analysis

Readig the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("Downloads/PumpFailure.csv")
```

Plotting the data

```python
def draw_line_plot(df, a):
    df_line = df.copy()
    df_line = df.set_index('timestamp')
    fig, ax = plt.subplots(figsize=(18,8))
    df_line[a].plot(ax=ax, color= "red").get_figure()
    ax.set_title("Pump failure")
    ax.set_ylabel("Value")
    ax.set_xlabel("Time")
 
draw_line_plot(df, 'value')
```

Checking for stationarity and Anomalies

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(df):
    
    rolmean = df.rolling(1000).mean()
    rolstd = df.rolling(1000).std()
    
    fig, ax = plt.subplots(2,1,figsize=(18,20))
    ax[0].plot(df,label='Original')
    ax[0].plot(rolmean, color='red', label='Rolling Mean')
    ax[0].plot(rolstd, color='black', label = 'Rolling Std')
    sns.histplot(data=df.rolling(1000).mean(), kde=True, bins = int(np.sqrt(df.rolling(200).mean().describe()[0])), ax=ax[1])
        
    dftest = adfuller(df, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    

    print ("p-value = " + str(dfoutput[1]))
    
test_stationarity(df['value'])
```

Univariate Analysis

```python
def draw_bar_plot(df):
    df_bar = df.copy()
    df_bar_t = df.copy()
    df_bar_t['value'] = np.log(df_bar['value'])

    
    df_bar = df_bar.set_index('timestamp')
    fig,ax = plt.subplots(1,2,figsize=(20,10))
    
    sns.histplot(data=df_bar, kde=True, bins = int(np.sqrt(df_bar.describe()['value'][0])), ax=ax[0])
    sns.histplot(data=df_bar_t, kde=True, bins = int(np.sqrt(df_bar.describe()['value'][0])), ax=ax[1])
    
    
    ax[0].set_title('No transformation')
    ax[0].axvline(x=df_bar.describe().values[4],color="green")
    ax[0].axvline(x=df_bar.describe().values[6],color="green")
    ax[0].axvline(x=df_bar.describe().values[1],color="gray")
    
    ax[1].set_title('ln transformation')
    ax[1].axvline(x=df_bar_t.describe().values[4],color="green")
    ax[1].axvline(x=df_bar_t.describe().values[4],color="green")
    ax[1].axvline(x=df_bar_t.describe().values[6],color="gray")
    print(df.describe())

draw_bar_plot(df)
```

```python
def draw_box_plot(df,a):
    df_box = df.copy()
    df_box_t = df.copy()
    df_box_t['value'] = np.log(df_box[a].values)
    df_box.reset_index(inplace=True)
    

    fig,ax = plt.subplots(1,2,figsize=(20,10))

    sns.boxenplot(data=df_box, y='value', ax=ax[0]) 
    sns.boxenplot(data=df_box_t, y='value', ax=ax[1]) 
    fig.savefig('box_plot.png')

draw_box_plot(df,'value')

```

Spectral Analysis

```python
from scipy import signal
from tensorflow.signal import rfft as fft
from tensorflow.signal import rfft as ifft

df_spec = df.copy()
#df_season = df_season.set_index(pd.to_datetime(df_season['timestamp']))
df_spec['timestamp'] = pd.to_datetime(df_spec['timestamp'])
df_spec = df_spec.set_index('timestamp')
df_spec.sort_index(inplace=True)
mean = df.describe()['value'][1]


fig,ax = plt.subplots(3,1,figsize=(20,10))

f, t, Sxx = signal.spectrogram(np.log(df_spec['value']), fs=1, nfft=5140)
ax[1].pcolormesh(t, f, np.log10(Sxx), shading='gouraud');
f, t, Sxx = signal.spectrogram((df_spec['value']), fs=1, nfft=5140)
ax[0].pcolormesh(t, f, np.log10(Sxx), shading ='gouraud');

fft = fft(((df['value'])-mean))
f_per_dataset = np.arange(0, len(fft))
n_samples_h = len(df['value'])
fivem_per_month = (60/5)*24*30
months_per_dataset = n_samples_h/(fivem_per_month)
f_per_month = f_per_dataset/months_per_dataset

ax[2] = plt.step(f_per_month, np.abs(fft))
ax[2] = plt.xscale('log')
ax[2] = plt.xticks([1,30,30/7], labels=['1/month', '1/day', '1/7 days'])
```

RNN forcasting

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

df_reg = df_out_z.copy()
df_reg['timestamp'] = pd.to_datetime(df_reg['timestamp'])
df_reg = df_reg.set_index('timestamp')
df_reg = df_reg.resample("H").mean()

n = len(df_reg)
train_df = df_reg[0:int(n*0.95)]
val_df = df_reg[int(n*0.95):]
      
sc = MinMaxScaler(feature_range=(0,1))
test_df_scale = sc.fit_transform(train_df)
   
time_step = 60
X_train = []
Y_train = []
m = len(test_df_scale)

for i in range(time_step,m):
    X_train.append(test_df_scale[i-time_step:i,0])
    Y_train.append(test_df_scale[i,0])
    
X_train, Y_train = np.array(X_train), np.array(Y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

dim_ent = (time_step,1)
dim_out = 1
na = 50
model = Sequential()
model.add(LSTM(units=na, input_shape=dim_ent))
model.add(Dense(units=dim_out))
model.compile(optimizer='rmsprop', loss='mse')
model.fit(X_train,Y_train,epochs=20,batch_size=32)
x_test = val_df.values
x_test = sc.transform(x_test)
X_test = []
for i in range(time_step,len(x_test)):
    X_test.append(x_test[i-time_step:i,0])
    
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

prediction = model.predict(X_test)
prediction = sc.inverse_transform(prediction)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,8))
ax.plot(val_df.values[0:len(prediction)], label='Real value')
ax.plot(prediction, color='gray', label='Forecast')
ax.legend()
```
